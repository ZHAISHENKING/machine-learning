# PCA

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
# PCA降维有如下两种方案
# 1. 确定要保留的数据特征个数 n_components设置为一个整数
# 2. 确定要保留的信息完整程度 n_components设置一个0-1小数
pca = PCA(n_components=0.95)
iris = load_iris()
train = pca.fit_transform(iris.data)
```

```python
plt.scatter(train[:,0],train[:,1],c=iris.target)
```

```python
# 查看特征选择是如何处理特征的
from sklearn.feature_selection import VarianceThreshold

#方差选择法，返回值为特征选择后的数据
#参数threshold为方差的阈值
train1 = VarianceThreshold(threshold=0.6).fit_transform(iris.data)
plt.scatter(train1[:,0],train1[:,1],c=iris.target)
display(train1.shape)
```

# LDA降维

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# n_components只能设置一个整数，表示要保留的特征维度
lda = LinearDiscriminantAnalysis(n_components=2)
train2 = lda.fit_transform(iris.data,iris.target)

plt.figure(figsize=(10,4))
axes1 = plt.subplot(1,2,1)
axes1.scatter(train2[:,0],train2[:,1],c=iris.target)
axes1.set_title('LDA')

axes2 = plt.subplot(1,2,2)
axes2.scatter(train[:,0],train[:,1],c=iris.target)
axes2.set_title('PCA')
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
# LDA分析后的评分
X_train,X_test,y_train,y_test = train_test_split(train2,iris.target,random_state=1)
knn = KNeighborsClassifier(n_neighbors=19)
knn.fit(X_train,y_train).score(X_test,y_test)
```

# 网格搜索

```python
功能：对指定算法参数进行调优，最终找到最优参数的算法模型
```

```python
# 借助GridSearchCV来对参数进行网格搜索
from sklearn.model_selection import GridSearchCV

parma_grid = {
    'n_neighbors':[3,5,7,9,11,13,15,17,19]
}
# estimator 指定要调整参数的模型
# param_grid 是一个字典 要调整的超参和参数值列表
clf1 = GridSearchCV(knn,parma_grid)

# 使用GridSearchCV对象进行训练
clf1.fit(X_train,y_train)

# 训练完成之后，就会得到一个最优解
best_knn = clf1.best_estimator_

# 使用最优解进行预测，查看评分
best_knn.score(X_test,y_test)
# 交叉验证查看算法稳定性（在样本集上进行处理）
```

```python
# 选择两个特征的评分
X_train,X_test,y_train,y_test = train_test_split(iris.data[:,:2],iris.target)
knn = KNeighborsClassifier()
knn.fit(X_train,y_train).score(X_test,y_test)
```

# 集成学习

功能：在多个算法模型中，找到最优算法

```python
from sklearn.ensemble  import VotingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
# 把所有要进行选择的算法模型存入此列表当中
estimators=[] #列表

# 构建要选择的模型
model1=LogisticRegression()
model2=SVC()
model3=DecisionTreeClassifier()

# 注意格式，把算法模型保存到列表中
estimators.append(("LogisticRegression",model1))
estimators.append(("SVC",model2))
estimators.append(("DecisionTreeClassifier",model3))

#算法集成工具，集成学习,看成一个机器学习模型
ensemble=VotingClassifier(estimators) 

# 使用LDA对鸢尾花数据集进行降维处理
train = LinearDiscriminantAnalysis(n_components=2).fit_transform(iris.data,iris.target)

# 确定样本集拆分的份数
kfold = KFold(n_splits=10)

# 使用交叉验证，来进行训练数据
result = cross_val_score(ensemble,train,iris.target,cv=kfold)
print(result.mean())
```

```python
ensemble.estimators
```

```python
from matplotlib.colors import ListedColormap

cmap = ListedColormap(['pink','cyan','orange'])
# ensemble对象可以直接用来训练和预测数据

# 使用集成学习算法模型绘制LDA降维后的分类边界

xmin,xmax = train[:,0].min()-0.5,train[:,0].max()+0.5
ymin,ymax = train[:,1].min()-0.5,train[:,1].max()+0.5
x = np.linspace(xmin,xmax,300)
y = np.linspace(ymin,ymax,300)

xx,yy = np.meshgrid(x,y)
x_test = np.c_[xx.ravel(),yy.ravel()]

ensemble.fit(train,iris.target)
y_ = ensemble.predict(x_test)
plt.scatter(x_test[:,0],x_test[:,1],c=y_,cmap=cmap)
plt.scatter(train[:,0],train[:,1],c=iris.target)
```
