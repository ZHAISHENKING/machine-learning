```python
X_train = np.load('./上午/code/x_train.npy')
X_test = np.load('./上午/code/x_test.npy')

y_train = np.load('./上午/code/y_train.npy')
y_test = np.load('./上午/code/y_test.npy')
```

```python
# 选择模型
# 决策树 支持向量机 逻辑斯提回归(主流的三种分类算法)
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 优化模型准确率有两条途径
# 1. 优化数据集
# 2. 调整参数
# 
knn = KNeighborsClassifier(n_neighbors=13)

# 训练数据
knn.fit(X_train,y_train)

# 预测数据
# knn 时间复杂度空间复杂度都比较高
y_ = knn.predict(X_test)
knn.score(X_test,y_test)
```

```python
# 算法评价
(y_==y_test).sum()/y_test.size
```
