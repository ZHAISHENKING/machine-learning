```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lenses = pd.read_csv('../DecitionTree/data/lenses.txt',sep='\t',header=None)
```

```python
# 先对分类标签进行数值映射
target = lenses[4]
map_target = {
    'no lenses':0,
    'soft':1,
    'hard':2,
    'young':0, 
    'pre':1, 
    'presbyopic':2,
    'myope':0, 
    'hyper':1,
    'no':0, 
    'yes':1
}
target = target.map(map_target)
```

```python
# 对数据特征进行替换操作，转换成数字
train = lenses.iloc[:,:3].copy()
for column in train:
    train[column] = train[column].map(map_target)
```

```python
# 确认数据特征都是可以运算的数字
train.dtypes
```

```python
# 分类问题 knn logistic decisiontree GaussionNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(train,target,test_size=0.2,random_state=3)
```

```python
print("KNN score is {}".format(KNeighborsClassifier(n_neighbors=9).fit(X_train,y_train).score(X_test,y_test)))
print("Logistic score is {}".format(LogisticRegression().fit(X_train,y_train).score(X_test,y_test)))
print("DecisionTree score is {}".format(DecisionTreeClassifier(max_depth=1).fit(X_train,y_train).score(X_test,y_test)))
print("Gaussian score is {}".format(GaussianNB().fit(X_train,y_train).score(X_test,y_test)))
```

```python
from sklearn.datasets import load_boston
```

```python
X,y = load_boston(return_X_y=True)
```

```python
train = DataFrame(X)
target = Series(y)
```

```python
y.min(),y.max()
```

```python
# 把波士顿房价预测改为分类问题
# 定义一个阈值，对样本标签进行映射处理
def transfom_2(x):
    if x <= 25:
        return 0
    else:
        return 1
target = target.map(transfom_2)
```

```python
# 对波士顿房价的特征进行归一化处理
for column in train:
    train[column] = train[column]/train[column].sum()
```

```python
X_train,X_test,y_train,y_test = train_test_split(train,target,test_size=0.2,random_state=2)
```
