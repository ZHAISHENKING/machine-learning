```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ablone = pd.read_table('abalone.txt',header=None)
ablone
```

```python
train = ablone.loc[:,0:7]
target = ablone[8]
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression,Ridge,Lasso
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 对数据样本集进行拆分
X_train,X_test,y_train,y_test = train_test_split(train,target,test_size=0.2,random_state=1)
```

```python
from sklearn.metrics import r2_score
```

```python
knn = KNeighborsRegressor(n_neighbors=19)
knn.fit(X_train,y_train)
knn_y_ = knn.predict(X_test)
print('KNN 模型预测评分是 {}'.format(r2_score(y_test,knn_y_)))
```

```python
line = LinearRegression(normalize=False)
line.fit(X_train,y_train)
line_y_ = line.predict(X_test)
print('LinearRegression 模型预测评分是 {}'.format(r2_score(y_test,line_y_)))
```

```python
ridge = Ridge(alpha=0.000001)
ridge.fit(X_train,y_train)
ridge_y_ = ridge.predict(X_test)
print('Ridge 模型预测评分是 {}'.format(r2_score(y_test,ridge_y_)))
```

```python
lasso = Lasso(alpha=0.0001)
lasso.fit(X_train,y_train)
lasso_y_ = lasso.predict(X_test)
print('Lasso 模型预测评分是 {}'.format(r2_score(y_test,lasso_y_)))
```
